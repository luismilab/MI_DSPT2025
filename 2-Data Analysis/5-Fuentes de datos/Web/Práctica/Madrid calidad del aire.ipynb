```python
import requests
import pandas as pd

url = "https://datos.madrid.es/egob/catalogo/212531-12751102-calidad-aire-tiempo-real.{responseContentType}"

try:
    response = requests.get(url)
    response.raise_for_status()
    data_json = response.json()

    # Accedemos a la lista de 'records' que contiene la informacin
    records = data_json.get('records')

    if records:
        df = pd.DataFrame(records)
        print(df)
    else:
        print("No se encontraron datos en la seccin 'records'.")

except requests.exceptions.RequestException as e:
    print(f"Error al conectar con la API: {e}")
except json.JSONDecodeError as e:
    print(f"Error al decodificar la respuesta JSON: {e}")
```
